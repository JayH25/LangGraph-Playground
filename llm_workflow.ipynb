```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint

from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
  repo_id="meta-llama/Meta-Llama-3-8B-Instruct",

  task = "text-generation"
)

model = ChatHuggingFace(llm = llm)
```

```python
class LLMstate(TypedDict):
  question: str
  answer: str
```

```python
def llm_qa(state: LLMstate)->LLMstate:
  q = state['question']
  prompt = f"Answer the following question {q}"
  ans = model.invoke(prompt).content
  state['answer'] = ans

  return state
```

```python
graph = StateGraph(LLMstate)


graph.add_node('llm_qa',llm_qa)

graph.add_edge(START,'llm_qa')
graph.add_edge('llm_qa',END)

workflow = graph.compile()
```

```python
initial_state={ 'question' : "who is so far the best or the popular and famous by fans opinion in adult film industry in female category?"}
final_state = workflow.invoke(initial_state)
print(final_state['answer'])
```
